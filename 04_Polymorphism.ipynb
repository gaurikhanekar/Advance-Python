# Polymorphism

### Polymorphism in Python is a core concept of object-oriented programming (OOP) that allows the same function or method name to behave differently depending on the object (class) calling it.

### In simple words: One name, many forms.

# 

```python
a = "Sarthak"
print(type(a))
print(len(a))
```

```python
b = [12, 13, 14, 15]
print(type(b))
print(len(b))
```

```python
c = (11, 12, 13)
print(type(c))
print(len(c))
```

```python
d = {"a": 1, "b": 2, "c": 3, "d": 4}
print(type(d))
print(len(d))
```

```python
d = [12, 13, 14, 15, 16]
print(type(d))
print(sum(d))
```

```python
e = (3, 4, 5, 6, 7, 8, 9)
print(type(e))
print(sum(e))
```

# Polymorphism in classes

```python

class India:

    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:

    def capital(self):
        print("Washington D.C. is capital of USA")
    
    def language(self):
        print("English is widely spoken in USA")
```

```python

class France:

    def capital(self):
        print("Paris is capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
print(type(c1))
c1.capital()
c1.language()
```

```python
c2 = USA()
print(type(c2))
c2.capital()
c2.language()
```

```python
c3 = France()
print(type(c3))
c3.capital()
c3.language()
```

```python
countries = [India(), USA(), France()]

for c in countries:
    print(type(c))
    c.capital()
    c.language()
    print("=" * 40)
```

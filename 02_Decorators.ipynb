# Write a decorator to welcome and thanks the user

```python
def welcome(func):
    def wrapper(*arg, **kwargs):
        print("Hello User!")
        res = func(*arg, **kwargs)
        print(f"Function Results : {res}")
        print("Thank you")
        return res
    return wrapper        
```

# If you want to apply decorator @decorator

```python
@welcome
def simple_interest(p: float, n: int, r: float) -> tuple:
    i = (p * n * r) / 100
    a = p + i
    return i, a
```

```python
i1, a1 = simple_interest(p=50_000, n=5, r=7.1)
```

# Write hypotenuse function with welcome decorator

```python
import math

@welcome
def hypotenuse(a: float | int, b: float | int) -> float:
    h = math.sqrt(a**2 + b**2)
    return h
```

```python
h1 = hypotenuse(a=3, b=4)
```

```python
h2 = hypotenuse(5, 6)
```

# Vowel counter welcome decorater

```python
@welcome
def vowel_counter(word: str):
    word = word.lower()
    count = 0
    for i in word:
        if i in "aeiou":
            count = count + 1
    return count
```

```python
c1 = vowel_counter("ETLHive")
```

```python
c2 = vowel_counter("Gauri Khanekar")
```

```python
c3 = vowel_counter("Rutika")
```

# Write a decorator to measure time required to execute a function

```python
import time
# Start the time counter
start = time.perf_counter() 
time.sleep(1)
print("Hello")
time.sleep(2)
print("World")
# Stop the time counter
stop = time.perf_counter()
# calculate elapsed time
elapsed = stop - start
print(f"Elaspsed time : {elapsed:.2f} sec")
```

```python
import time

def measure_time(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        res = func(*args, **kwargs)
        stop = time.perf_counter()
        print(f"Results : {res}")
        elapsed = stop - start
        print(f"Time elapsed : {elapsed:.2f} sec")
        return res
    return wrapper
```

```python
@measure_time
def cube(num: int) -> int:
    time.sleep(5)
    return num ** 3
```

```python
c1 = cube(12)
```

```python
c2 = cube(123)
```

# Write a function to calculate product od=f given number in a list 

```python
@measure_time
def product_nums(a: list[int]) -> int:
    # Intialize product to 1
    p = 1
    # Apply for loop on the list a
    for i in a:
        time.sleep(1)
        print(f"Multuplying : {i}")
        p = p * i
    return p
```

 # From data anaysis point of view write a decorator function that validates type of data and maximum limit 

```python
def data_validator(datatype, max_length):
    def decorator(func):
        def wrapper(data):
            if not isinstance(data, datatype):
                print(f"Invalid datatype expected {datatype} got type : {type(data)}")
                return None
            if len(data) > max_length:
                print(f"Max length exceeded {max_length}")
                return None
            print(f"Valid datatype : {datatype} and under {max_length} limit")
            return func(data)
        return wrapper
    return decorator
```

```python
@ data_validator(datatype=list, max_length=5)
def average_nums(a : list):
    return sum(a) / len(a)
```

```python
a = [1, 2, 3, 4]
avg1 = average_nums(a)
print(avg1)
```

```python
b = {"a": 3, "b": 4, "c": 5}
avg2 = average_nums(b)
print(avg2)
```

```python
c = [3, 11, 12, 13, 14, 15, 16]
avg3 = average_nums(c)
print(avg3)
```

```python

@data_validator(datatype=dict, max_length=6)
def analyze_marks(marks: dict) -> dict:
    marks_values = marks.values()
    max_marks = max(marks_values)
    min_marks = min(marks_values)
    avg_marks = sum(marks_values) / len(marks_values)
    return {
        "min_marks": min_marks,
        "avg_marks": avg_marks,
        "max_marks": max_marks
    }
```

```python
r1 = analyze_marks(
    {
        "Rahul": 86,
        "Raman": 75,
        "Aditi": 63,
        "Sarthak": 93
    }
)
print(r1)
```

```python
m = (73, 85, 96, 101, 105)
r2 = analyze_marks(m)
print(r2)
```

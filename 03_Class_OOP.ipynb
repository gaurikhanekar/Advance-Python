# Create a bank account class

```python
class BankAccount:
    
    def __init__(self, ac_no: int, name: str, balance: float = 0.0):
        self.ac_no = ac_no
        self.name = name
        self.balance = balance
        if balance < 0:
            raise ValueError("Negative Balance is Invalid")
        
    def __str__(self):
        return f"BankAccount(ac_no = {self.ac_no}, name = {self.name})"
        
    def get_balance(self):
        print(f"Account No : {self.ac_no}, Balance : {self.balance} INR")

    def deposit(self, amt: float):
        if amt <= 0:
            raise ValueError("Amount cannot be negative or zero")
        self.balance = self.balance + amt

    def withdraw(self, amt: float):
        if amt <= 0:
            raise ValueError("Amount cannot be negative or zero")
        if self.balance < amt:
            raise ValueError("Insufficient balance")
        self.balance = self.balance - amt

    def transfer(self, amt: float, other_ac: 'BankAccount'):
        self.withdraw(amt)
        other_ac.deposit(amt)
```

```python
import math

class Triangle:
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    # Check if given sides form a valid triangle
    def is_valid(self):
        return (
            self.side1 + self.side2 > self.side3,
            self.side2 + self.side3 > self.side1,
            self.side1 + self.side3 > self.side2
        )

    # Perimeter of the triangle
    def perimeter(self):
        if not self.is_valid():
            return "Invalid triangle. Cannot compute perimeter."
        return self.side1 + self.side2 + self.side3

    # Area of the triangle using Heron's Formula
    def area(self):
        if not self.is_valid():
            return "Invalid triangle. Cannot compute area."
        
        s = self.perimeter() / 2
        area = math.sqrt(
            s * (s - self.side1) * (s - self.side2) * (s - self.side3)
        )
        return area
    
t = Triangle(3, 4, 5)

print("Side1:", t.side1)
print("Side2:", t.side2)
print("Side3:", t.side3)

print("\nIs this a valid triangle?", t.is_valid())
print("Perimeter:", t.perimeter())
print("Area:", t.area())    
```

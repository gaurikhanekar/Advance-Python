# Dunder Functionm
1. attributes - constructor function \_\_init\_\_

```python
class Person:
    
    def __init__(self, name: str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    def intro(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as {self.occupation}")
```

```python
p1 = Person(name="Sarthak", age=23, gender="male", occupation="Sales Engg.")
print(p1)
```

# Create a person2 class with string representation __str()__

```python
class Person2:

    def __init__(self, name:str, age: int, gender: str, occupation: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    def __str__(self):
        return f"Person(name = {self.name}, age = {self.age}, gender = {self.gender}, occupation = {self.occupation})"

    def talk(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as a {self.occupation}")        
```

```python
pe1 = Person2(name="Rahul", age=25, gender="male", occupation="CA")
pe1
```

```python
print(pe1)
```

```python
pe1.name
```

```python
pe1.gender
```

```python
pe1.occupation
```

```python
pe1.talk()
```

# Dataclasses in python

```python
from dataclasses import dataclass
from typing import Literal
```

```python
@dataclass
class Person3:
    name: str
    age: int
    gender: Literal["male", "female"]
    occupation: str

    def speak(self):
        print(f"My name is {self.name}")
        print(f"I am {self.age} years old")
        print(f"I am {self.gender}")
        print(f"I work as a {self.occupation}")
```

```python
p3 = Person3(name="Sarthak", age=28, gender="male", occupation="HR")
type(p3)
```

```python
def division2(num: float, den: float) -> float:
    try:
        return num / den
    except Exception as e:
        print(f"Exception occured: {e}")    
```

# Exception Handling method 2

```python
def division3(num: float, den: float) -> float:
    try:
        return num / den
    except ZeroDivisionError:
        print("Denominator of Division cannot be 0")
    except TypeError:
        print("Cannot divide string and number, please enter number only")        
```

```python
r1 = division3(num=35, den=5)
print(r1)
print("Done")
```

```python
r2 = division3(num="abc", den=3)
print(r2)
print("Doe")
```

```python
r3 = division3(num=12, den=0)
print(r3)
print("Done")
```

```python
r4 = division3("a", "b")
print(r3)
print("Done")
```

# Other blocks try, except, else, finally
1. try - Any code that you wn=ant to execute
2. except - will trigger if any exception(error) occures.
3. else - will trigger if try blosck execute successfully.
4. finally - This blosck will always executed

```python
import math 
math.sqrt(25)
```

```python
math.sqrt(2)
```

```python
math.sqrt(3.5)
```

```python
math.sqrt(-2)
```

```python
math.sqrt("abc")
```

```python
def square_root(num: float | int) -> float:
    try:
        s = math.sqrt(num)
    except Exception as e:
        print(f"Exception occured : {e}")
    else:
        print(f"Square root of {num} is {s:.4f}")
    finally:
        print(f"Finally Block will Always execute")            
```

```python
square_root(25)
```

```python
square_root(-2)
```

```python
square_root("abc")
```

```python
from urllib.request import urlretrieve

def download_file(url: str):
    filename = url.split("/")[-1]
    print(f"Downloading {filename} ...")
    urlretrieve(url, filename)
    print(f"{filename} download complete!")
```

```python
download_file(
    url="https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/daily_revenue.csv"
    )
```

```python
def download_exception(url: str):
    try:
        filename = url.split("/")[-1]
        print(f"Download started for {filename} ...")
        urlretrieve(url, filename)
        print(f"Download {filename} complete")
    except  Exception as e:
        print(f"Exception occured : {e}")    
```

```python
download_exception(
    url="https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/daily_revenue.csv"
)
```

```python
download_exception(url="bcd")
```
